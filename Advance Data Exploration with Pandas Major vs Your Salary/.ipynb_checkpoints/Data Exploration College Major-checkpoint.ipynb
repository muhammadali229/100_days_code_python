<p>College degrees are very expensive. But, do they pay you back? Choosing Philosophy or International Relations as a major may have worried your parents, but does the data back up their fears? PayScale Inc. did a year-long survey of 1.2 million Americans with only a bachelor's degree. We'll be digging into this data and use Pandas to answer these questions:</p>



<ol>
    <li>Which degrees have the highest starting salaries?</li> 
    <li>Which majors have the lowest earnings after college?</li>
    <li>Which degrees have the highest earning potential?</li>
    <li>What are the lowest risk college majors from an earnings standpoint?</li>
    <li>Do business, STEM (Science, Technology, Engineering, Mathematics) or HASS (Humanities, Arts, Social Science) degrees earn more on average?</li>
</ol>



```python
# import libraries
import pandas as pd
```

```python
pd.options.display.float_format = '{:,.2f}'.format
```

### Find College Major with Highest Starting Salaries

```python
df.loc[df['Starting Median Salary'].idxmax(), 'Undergraduate Major']
```

```python
df[df['Starting Median Salary'] == df['Starting Median Salary'].max()]['Undergraduate Major']
```

### What college major has the highest mid-career salary? How much do graduates with this major earn? (Mid-career is defined as having 10+ years of experience)

```python
df.loc[df['Mid-Career Median Salary'].idxmax(), ['Undergraduate Major', 'Mid-Career Median Salary']]
```

### Which college major has the lowest starting salary and how much do graduates earn after university?

```python
df.loc[df['Starting Median Salary'].idxmin(), ['Undergraduate Major', 'Starting Median Salary']]
```

### Which college major has the lowest mid-career salary and how much can people expect to earn with this degree?

```python
df.loc[df['Mid-Career Median Salary'].idxmin(), ['Undergraduate Major', 'Mid-Career Median Salary']]
```

### Majors with the Most Potential vs Lowest Risk

```python
spread = df['Mid-Career 90th Percentile Salary'] - df['Mid-Career 10th Percentile Salary']
df.insert(1, 'Spread', spread)
```

```python
df.sort_values('Spread')[['Undergraduate Major', 'Spread']].head()
```

```python
df.sort_values('Mid-Career 90th Percentile Salary', ascending=False)[[
    'Undergraduate Major', 'Mid-Career 90th Percentile Salary'
]].head()
```

```python
df.sort_values('Spread', ascending=False)[['Undergraduate Major', 'Spread']].head()
```

```python
df.sort_values('Mid-Career Median Salary', ascending=False)[['Undergraduate Major', 'Mid-Career Median Salary']].head()
```

### Which category of degrees has the highest average salary? Is it STEM, Business or HASS (Humanities, Arts, and Social Science)? 

```python
df.groupby('Group').size()
```

### Find the average salary by group? 

```python
df.groupby('Group').mean()
```
